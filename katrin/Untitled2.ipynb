```python
# spojky AND / OR : když jde o slovíčkaření ...
# nevím co o tom napsat.. závorkami se dělá hierarchie,
# a je to jednoduchý AND, OR, XOR .... (popřípadě tě s tím seznámím)
# u některých lze používat klíčová slova, jinak symboly

for i in range(0,2):
  for j in range(0,2):
    print(f" {int(i)} OR {int(j)} = {int(i | j)}")

print("\n")

for i in range(0,2):
  for j in range(0,2):
    print(f" {int(i)} AND {int(j)} = {int(i & j)}")

print("\n")

for i in range(0,2):
  for j in range(0,2):
    print(f" {int(i)} XOR {int(j)} = {int(i ^ j)}")
```

# 5. Manipulace se soubory

```python
# je to velmi jednoduché, jen pozor na přístupová práva !!
# instanciace objektu: f = open("jmeno_souboru", opravneni)
# oprávnění:
# "r" - read, pouze čtení
# "w" - write, giga gauner, smaže obsah souboru a když neexistuje, vytvoří jej
# "a" - append, přidání na konec

f = open("milan.txt", "w")
f.write("Ahoj, jsem Kamil!")
# bývá dobrým zvykem soubory pak co nejrychleji zavřít
f.close()

f = open("milan.txt", "r")
print("První zápis:")
print(f.read())
f.close()

f = open("milan.txt", "a")
f.write("\nAhoj, jsem Kamil!")
f.close()

f = open("milan.txt", "r")
print("\nDruhý zápis:")
print(f.read())
f.close()

f = open("milan.txt", "r")
print("\nJednotlivé řádky:")
print(f.readlines())
f.close()
```
